```python
import requests
```

```python
#URL de la API
url = "https://api-intranet-jira.portinos.com/v1/api/get-time-issue/DEVLAR-1738"
```

```python
#parametros
parametros = {
    "data": "data",
    "userName": "userName",
    "hs": "hs"
}
```

```python
#solicitud Get
response = requests.get(url, params=parametros)
```

```python
#verifico si es exitosa la respuesta
if response.status_code == 200:
    #respuesta json
    datos = response.json()
    print(datos)
else:
    print(f"Error de la solicitud: {response.status_code}")
```
